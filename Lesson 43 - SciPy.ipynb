```python
# solving an equation

import numpy as np
def eqn(x):
    return x * np.cos(x) + np.sin(x) / x

print(eqn(-1))
print(eqn(1))
```

```python
# plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(-10,10)
y = eqn(x)
ax = plt.gca()
ax.plot(x,y)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
plt.show()
```

```python
# solve
from scipy.optimize import root
root(eqn,-1)
```

```python
root(eqn,1)
```

```python
# minimise
from scipy.optimize import minimize
def eqn(x):
    return x**4 + x**3 + x**2 + x + 1

print(eqn(-1))
print(eqn(0))
print(eqn(1))
```

```python
%matplotlib inline
x = np.linspace(-2,2)
y = eqn(x)
ax = plt.gca()
ax.plot(x,y)
ax.spines['left'].set_position('center')
plt.show()
```

```python
minimize(eqn,0)
```
