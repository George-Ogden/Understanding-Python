```python
import yaml
```

```python
with open("elements.yaml") as file:
    elements = yaml.load(file,Loader=yaml.FullLoader)
```

```python
elements
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(elements)
df.set_index("name",inplace=True)
df
```

```python
df.columns
```

```python
df.drop(["cpkHexColor","standardState","bondingType"],inplace=True,axis=1)
df
```

```python
import re
```

```python
def convert(name):
    return " ".join(re.split(r"(?<=.)(?=[A-Z])",name[0].upper() + name[1:]))
```

```python
convert("electronConfiguration")
```

```python
df.rename(columns=dict([[name,convert(name)] for name in df.columns]),inplace=True)
df
```

```python
df.rename(columns={"Van Der Waals Radius":"Van der Waals Radius"},inplace=True)
df
```

```python
df[~df["Year Discovered"].isnull()].sort_values(["Year Discovered"])
```

```python
json = df.to_dict("index")
json
```

```python
with open("clean.yaml","w") as file:
    yaml.dump(json,file)
```
